```python
import tkinter as tk
from tkinter import ttk

def on_combobox_select(event):
    selected_option = combobox.get()
    if selected_option == "Yes":
        response_label.config(text="You chose Yes!")
    elif selected_option == "No":
        response_label.config(text="You chose No!")

root = tk.Tk()
root.geometry("300x150")
root.title("Choose an Option")

options = ["Yes", "No"]

# Tworzymy etykietę informacyjną
info_label = tk.Label(root, text="Choose an option:")
info_label.pack(pady=10)

# Tworzymy ComboBox z dostępnymi opcjami
combobox = ttk.Combobox(root, values=options)
combobox.pack()

# Ustalamy domyślną opcję (jeśli chcesz)
combobox.set(options[0])

# Tworzymy etykietę na odpowiedź
response_label = tk.Label(root, text="", font=("Helvetica", 12))
response_label.pack(pady=10)

# Powiązujemy zdarzenie wyboru z funkcją on_combobox_select
combobox.bind("<<ComboboxSelected>>", on_combobox_select)

root.mainloop()
```
